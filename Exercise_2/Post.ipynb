```python
# Importing libraries to complete analysis
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading in the example2.rdf file line by line and storing it in log_file
rdf_file = []
file1 = open('example2.rdf', 'r')
count = 0
while True:
    count +1
    line = file1.readline()
    if not line:
        break
    rdf_file.append(line)
file1.close()
```

```python
startpoint = 3
tmp = rdf_file[startpoint].split()
bins = int(tmp[1])
rdf_dictionary = {}

for k in range(0,20):
    data_start = k*bins + startpoint + 1 + k
    data_end = (k+1)*bins + data_start
    rdf_np = np.array(rdf_file[data_start:data_end])
    output = np.zeros((bins,3))
    for i in range(0,bins):
        tmp = rdf_np[i].split()
        for j in range(0, 3):
            output[i][j] = tmp[j]
    rdf_dictionary[k] = pd.DataFrame(output, columns = ['BinNumber', 'RadialDistance', 'Count'])
```

```python
fig = plt.figure(figsize = (6.4,4.8))
ax = plt.axes((0.1,0.15,0.8,0.8))
ax.plot(rdf_dictionary[0].RadialDistance,rdf_dictionary[0].Count,label ='Step = 0')
ax.plot(rdf_dictionary[1].RadialDistance,rdf_dictionary[1].Count,label ='Step = 500')
ax.plot(rdf_dictionary[2].RadialDistance,rdf_dictionary[2].Count,label ='Step = 1000')
ax.legend(facecolor='white',edgecolor='black',framealpha=1,fontsize=14)
ax.set_xlabel('Radial Distance [LJ units]',fontsize=14)
ax.set_ylabel('Count',fontsize=14)
ax.set_title('Exercise 2',fontsize=14)
plt.savefig('Exercise_2.jpg', format='jpg',dpi=300,bbox_inches='tight')
```
